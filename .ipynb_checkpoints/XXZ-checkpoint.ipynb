```julia
using LinearAlgebra
using SparseArrays

N = 8
data_b=zeros(Int64,2^N)
data_c=zeros(Int64,2^N)
data_d=zeros(Int64,N+1)

for i in 0:2^N-1
    M = 0
    a = digits(i,base = 2,pad = N )

    for k in 1:N
        M = a[k] + M
            
    end    
    data_b[i+1] = M 
    data_d[M+1]+=1
    data_c[i+1]=data_d[M+1]
    
end
```

```julia
Ev = zeros(2^N)
VEC = spzeros(2^N,2^N)
h=1
λ=0.0000001
q = 0.001
Δ = (q + q^-1)/2 
Γ = (q - q^-1)/2 
J=0

for i in 0:N
    A = findall(==(i),data_b)  
    B = length(A)
    F= zeros(Int32, B)
    H = zeros(B,B)
    
    for k in 1: B        
        C =  digits(A[k]-1,base = 2,pad = N )      
        H[k,k]=h*(N-2*i)  
        E = reverse(C)   
        Ww=0
        
        for i in 0:N-1   
        Ww = (2^i)*E[i+1]+Ww
            
        end      
        F[k]=Ww+1
                    
        for o in 1:N-1         
            O = C[o] +C[o+1]
            H[k,k]=  H[k,k]+ λ*Δ*(-1)^O　
            
            if O==1 
                D = reverse(C,o,o+1)
                W = 0
                
                for i in 0:N-1
                    W = W + (2^i)*D[i+1]
                    
                end
                W=W+1
                l = findall(==(W),A) 
                H[l[1],k]= H[l[1],k]+2*λ
            
            end
        end
        if  C[1]==0 && C[N]==1
            H[k,k]=H[k,k]+2*Γ*λ
        
        elseif   C[1]==1 && C[N]==0
            H[k,k]=H[k,k]-2*Γ*λ
                
        end
    end
E =  eigen(H)  
    
    for  k in 1:B
        Ev[J+k]  =  E.values[k]
                        
        for l in 1:B
            VEC[F[k],J+l] = E.vectors[k,l]
                      
        end
    end
    J+=B
    
end
```

```julia
t= 0.5

function Deigen(N,a,dt,LorR) 
    tl = t + dt/2
    tr = t - dt/2
    data_x= spzeros(2^N,2^N)
    data_z = spzeros(2^N,2^N)
    data_S =spzeros(2^N,2^N)
    data_SX = spzeros(2^N,2^N)
    data_b=zeros(Int64,2^N)
    
    for i in 0:2^N-1
        M = 0
        aa= digits(i,base = 2,pad = N)

        for k in 1:N
            M = aa[k] + M
        
        end
        data_b[i+1] = M
    
    end
    
    if  LorR==0
        T=tl
        position=1
        
    else LorR==1
        T=tr
        position=N
                
    end
    J=0
    K=0

    for i in 0:N 
        A = findall(==(i),data_b) 
        B = length(A)
    
        for k in 1: B
            C =  digits(A[k]-1,base = 2,pad = N )  
            D =  digits(A[k]-1,base = 2,pad = N )
            O = findall(==(i-1),data_b)
            O1 = length(O)
            Q = findall(==(i+1),data_b)
            Q1 = length(Q)       
            D = reverse(D)
            W=1
        
            for i in 0:N-1
                W = W + (2^i)*D[i+1]
            
            end
            if C[position]==1　
                C[position]=0
                E = reverse(C) 
                Ww=1
        
                for i in 0:N-1
                    Ww = (2^i)*E[i+1]+Ww
                    
                end           
                for s in 1:B     
                    v1 =VEC[:,J+s]    　
                    v2 = adjoint(v1)
                    data_z[J+s,J+s]= (a/2)*(1-T)*v2[W]*v1[W]+data_z[J+s,J+s]  
                    
                end        
                for  (b, c) in zip([ J+1-O1,J+B+1],[J,J+B+Q1])      
                    for m in b:c                     
                        if VEC[Ww,m]!=0                        
                            v2 = adjoint(VEC[:,m])　
                    
                            for s in 1:B
                                v1 =VEC[:,J+s]   
                                v3 =adjoint(v1)                 
                                data_x[J+s,m]= -v2[Ww]*v1[W]+data_x[J+s,m]
                                data_S[m,J+s]= -(a/2)*(1+T)*v2[Ww]*v1[W]+data_S[m,J+s]
                                
                            end                                                
                        end                
                    end                
                end       
                
            else C[position]==0 
                C[position]=1                
                E = reverse(C)
                Ww=1
        
                for i in 0:N-1                 
                    Ww = (2^i)*E[i+1]+Ww
                
                end        
                for s in 1:B            
                    v1 =VEC[:,J+s]    
                    v2 = adjoint(v1)
                    data_z[J+s,J+s]= (a/2)*(1+T)*v2[W]*v1[W]+data_z[J+s,J+s]     
                
                end         
                for  (b, c) in zip([ J+1-O1,J+B+1],[J,J+B+Q1])     
                    for m in b:c                  
                        if VEC[Ww,m]!=0                                       
                            v2 = adjoint(VEC[:,m])　                     
                            for s in 1:B                    
                                v1 =VEC[:,J+s]    
                                data_x[J+s,m]= -v2[Ww]*v1[W]+data_x[J+s,m]
                                data_S[m,J+s]= -(a/2)*(1-T)*v2[Ww]*v1[W]+data_S[m,J+s]                           
                            end                                                  
                        end       
                    end              
                end         
            end

            K+=1    
            if K==B         
                for s in 1:B
                    for  (b, c) in zip([ J+1-O1,J+B+1],[J,J+B+Q1])       
                        for m in b:c             
                            data_SX[J+s,m] = data_S[m,J+s]*data_x[J+s,m]
                        
                        end                  
                    end               
                end         
                J+=B
                K=0
        
            end     
        end
    end    
    DLorR=2*(data_SX-data_z)
    
    return DLorR
end
```

```julia
dt=0.2
tl = t + dt/2
tr = t - dt/2
DL=Deigen(N,1,dt,0)
DR=Deigen(N,1,dt,1)
```

```julia
D = spzeros(2^N,2^N)
D = DL+DR
I4=zeros(2^N)

for i in 1:2^N   
    I4[i] = 1

end
A = D'*D
p0 = (1/2^N)*I4
b = -A*p0 
```

```julia
n = 2^N
function multAvec(A::SparseMatrixCSC{Float64, Int64}, vec::Array{Float64,1})
    return A*vec
end
multA(vec::Array{Float64,1}) = multAvec(A,vec)

#CG法
function CG(b::Array{Float64, 1}, x₀::Array{Float64, 1}, maxiter::Int64, ϵ::Float64)
    #探索方向ベクトル, 残差ベクトル, 行列-ベクトル積, 近似解
    r = b- multA(x₀)
    p = r
    Ap = multA(p)
    x = x₀

    #残差ベクトルのノルム
    normr_vec = zeros(maxiter+1)
    normr = norm(r)
    normr_sq = normr^2
    normr_vec[1] = normr
    if normr < ϵ
        return x, normr_vec[1:k+1]
    end

    #メインループ
    for k in 1:maxiter
        #更新
        α = normr_sq/dot(p,Ap)
        x = x + α*p
        r = r - α*Ap
        normr = norm(r)
        normr_vec[k+1] = normr

        #収束のチェック
        if normr < ϵ
            return x, normr_vec[1:k+1]
        end

        β = normr^2/normr_sq
        p = r + β*p
        Ap = multA(p)
        normr_sq = normr^2
    end
    return x, normr_vec
end

x₀ = zeros(n)
maxiter = 1000
ϵ = 1e-6
x, normr_vec = CG(b, x₀, maxiter, ϵ)

Z = x+p0
Z1=0

for i in 1:2^N
    
    Z1 = Z[i]+Z1
    
end

E0 = (x+p0)/Z1

/
```

```julia
Z = [1 0
0 -1]
tj1 = zeros(N)
tj2 = zeros(N+2)
tj2[1]=tl
tj2[N+2]=tr

for i in 0:N-1 
    ZW1 = 0
    ZJ = 0
    Il = Matrix{Int64}(I,2^i,2^i)
    Ir = Matrix{Int64}(I,2^(N-1-i),2^(N-1-i))
    Zj = kron(Il,Z,Ir)
    
   for k in 1:2^N   
        v1 =VEC[:,k]    　
        v2 = adjoint(VEC[:,k])　    
        ZJ1 = (v2*Zj*v1)
        ZJ1 = ZJ1*E0[k]
        ZW1 = ZJ1 + ZW1
    
    end  
    l = 3   
    v1 = VEC[:,l]
    v2 = adjoint(v1)  
    ZJ = (v2*Zj*v1)  
    tj1[i+1] = -ZJ
    tj2[i+2] = -ZW1
    
end
```
